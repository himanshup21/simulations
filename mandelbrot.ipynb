```python
import numpy as np
from matplotlib import pyplot as plt

def mandelbrot(re, im):
    c = complex(re,im)
    z = 0 + 0j
    
    for i in range(100):
        z = z**2 + c
        if abs(z) > 2:
            return (i)

    return (i)


rows = 2000
cols = 2000

points = np.zeros([rows,cols])

for r_index, re in enumerate(np.linspace(-2.5,0.75,rows)):
    for c_index, im in enumerate(np.linspace(-1.5,1.5,cols)):
        points[r_index,c_index] = mandelbrot(re,im)
        
fig = plt.figure(figsize=(15,15))
plt.imshow(points.T, cmap='RdGy', extent = [-2.5,0.7,-1.5,1.5])

```
