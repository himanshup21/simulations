```python
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import clear_output
import time

def mandelbrot(re, im):
    c = complex(re,im)
    z = 0 + 0j

    for i in range(50):
        z = z**2 + c
        if abs(z) > 2:
            return (0)
            break

    return (1)

rows = 1000
cols = 1000

points = np.zeros([rows,cols])

n = 101
#time.sleep(5)
for i in range(n):
    for r_index, re in enumerate(np.linspace(-1.25 - 0.12*i/(n-1),-0.75 - 0.5*i/(n-1),rows)):
        for c_index, im in enumerate(np.linspace(-0.25 + 0.19*i/(n-1),0.25 - 0.19*i/(n-1),cols)):
            points[r_index,c_index] = mandelbrot(re,im)
    
    fig = plt.figure(figsize=(20,20))
    plt.imshow(points.T, cmap='binary')#,extent = [-1.26 - 0.11*i/n,-0.75 - 0.505*i/n,-0.25 + 0.185*i/n,0.25 - 0.185*i/n])
    clear_output(wait=True)
    plt.show()
```
